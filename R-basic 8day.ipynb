# 11장. 효율적인 프로그래밍

+ 사용자 정의 함수는 반복되는 프로그램 코드를 복사하여 붙여넣을 필요가 없고 수정이 필요하다면 사용자 정의 함수에서 한 번만 수정하면 된다
+ 사용자 정의 함수는 한번 만들어 놓으면 필요할 때마다 불러서 사용 가능하고 다른 프로그램에서도 사용자 정의 함수를 사용하여 프로그램할 수 있다.
+ 사용자 정의 함수를 이용하여 작업 과정을 나누어서 프로그램한다면 훨씬 편리하게 프로그램할 수 있다


+ 사용자 정의 함수는 매개변수, 지역변수, 상위변수, 전역변수로 나뉜다
+ 매개변수는 함수를 사용할 때, 함수 사용자가 함수에 제공하는 값을 말한다
+ 지역변수는 함수 내에서만 기억공간을 확보하여 데이터를 저장하고 있지만, 함수를 벗어나면 사라지는 변수이다
+ 전역변수는 함수 외부에 기억공간이 확보되어 있어 함수를 벗어나도 기억 공간이 유지되는 변수이다 ex) a <<- 10
+ 상위변수는 우변에만 해당하는 개념으로써 함수 내의 지역에 해당 변수가 없을 때 상위 지역의 변수를 참조하는 것이다

```r
calcscore <- function(){
    totvalue <- sum(score1)
    avavalue <- mean(score1)
    cat("성적 합계 :", totvalue,"성적 평균 :",avavalue,"\n")
    score1[2] <<- 95
}

score1 <- c(70, 80, 90, 60)
calcscore()
```

```r
calcscore <- function(s){
    totvalue <- sum(s)
    avavalue <- mean(s)
    cat("성적 합계 :", totvalue,"성적 평균 :",avavalue,"\n")
}

score1 <- c(70, 80, 90, 60); calcscore(score1)
score2 <- c(80, 75, 65, 96); calcscore(score2)
score3 <- c(50, 92, 75, 80); calcscore(score3)
```

```r
calcscore <- function(k, e, m){
    totvalue <- k + e + m
    avavalue <- totvalue / 3
    cat("성적 합계 :", totvalue,"성적 평균 :",avavalue,"\n")
}

kor <- c(70, 80, 90, 60);
eng <- c(80, 75, 65, 96); 
mat <- c(50, 92, 75, 80);
calcscore(kor, eng, mat)
```

```r
calcscore2 <- function(s){
    totvalue <- sum(s)
    avavalue <- mean(s)
    result <- c(totvalue, avavalue)
    return(result)
    cat("성적 합계 :", totvalue,"성적 평균 :",avavalue,"\n")
}

score1 <- c(70, 80, 90, 60);
result1 <- calcscore2(score1)
score2 <- c(80, 75, 65, 96);
result2 <- calcscore2(score2)

if(result1[2] > result2[2]){
    print("1반이 2반보다 더 성적이 우수합니다")
} else {
    print("2반이 1반보다 더 성적이 우수합니다")
}
```

### 사용자 패키지 함수 만들기

+ 사용자 패키지 함수를 만드려면 cran사이트에서 Rtools를 설치해야 한다

```r
fourCalcFun <- function(a,b){
  totvalue <- a + b; minusvalue <- a - b
  mulvalue <- a * b; divvalue <- a / b
  cat(a, "+", b, "=", totvalue,"\n")
  cat(a, "-", b, "=", minusvalue,"\n")
  cat(a, "*", b, "=", mulvalue,"\n")
  cat(a, "/", b, "=", divvalue,"\n")
}

divfun <- function(a,b){
  quotient <- a / b
  remainder <- a %% b
  cat(a, "/", b, "=", quotient, "\n")
  cat(a, "%%", b, "=", remainder, "\n")
}
```

+ 새로운 프로젝트를 만든 후 이 함수들을 패키지로 만들어 사용할 수 있다

```r
makescoredata <- function(){
    number <- seq(1:5); name <- c("송중기","전지현","송혜교","김태희","정우성")
    kor <- c(58, 63, 85, 69, 90); eng <- c(100, 95, 63, 80, 32)
    mat <- c(98, 45, 56, 89, 60); grade <- c("B","D","A","C","D")
    stuinfo <<- data.frame(number, name, kor, eng, mat, grade)
    colnames(stuinfo) <- c("번호","이름","국어","영어","수학","등급")
    cat("\n성적 데이터 :\n"); print(stuinfo); cat("\n");
}

calcsummean <- function(){
    spos <- 3; tpos <- 5;
    totscore <- apply(stuinfo[,c(spos:tpos)],1,sum)
    avgscore <- apply(stuinfo[,c(spos:tpos)],1,mean)
    resscore <<- data.frame(stuinfo[2],totscore,avgscore)
    cat("\n세과목 합계의 평균 :\n"); print(resscore, digits = 4)
}

overbase70 <- function(){
    basescore <- 70
    stuoverbase <- resscore[resscore$avgscore > basescore, ]
    cat("\n", basescore, "점 이상인 학생은 : \n"); print(stuoverbase, digits = 4)
}

makescoredata()
calcsummean()
overbase70()
```
