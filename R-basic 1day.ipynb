# 2장. R에서의 데이터 유형과 연산

### 데이터와 변수

+ 컴퓨터 프로그램 언어에서는 특정 값이 지정된 메모리의 위치를 편리하게 관리하기 위해서 메모리의 위치에 이름을 붙여 관리한다
+ 컴퓨터 언어에서는 이 이름을 변수라고 부르며 일반적으로 변수는 메모리의 특정 공간을 가리킨다고 할 수 있다
+ 변수 이름은 알파벳,숫자,언더바,마침표로 구성하며 첫글자는 반드시 알파벳 또는 마침표로 시작하여야 하며 마침표 뒤에는 숫자가 올 수 없다
+ 알은 대소문자를 구분하여 사용하므로 'num'과 'Num'은 다른 변수이다
+ 'print'와' 'sum'과 같은 함수 명을 변수로 사용할 경우, 해당 함수를 사용할 수 없으므로 가능한 사용하지 않는것이 바람직하다
+ 할당연산자는 <- 으로 하며 등호나 오른쪽 화살표는 거의 쓰지않는다

### 데이터의 유형

+ 컴퓨터의 메모리는 0과 1만의 정보를 저장할 수 있다.
+ 이진 값을 읽어가는 컴퓨터의 명령어에 따라 정보가 달라지며 숫자와 문자를 구분할 수 있게 된다
+ 데이터 유형이란 데이터가 의미하는 바가 유지되도록 이진수로 변환하고 해석하는 방법을 지정하는 것
+ R언어에서 사용하는 주요 데이터 유형은 숫자,문자,논리,날짜 및 시간형이 있다.

### 숫자형과 문자형

+ R에서는 실수형과 정수형을 따로 구분하지는 않음
+ 학번같은 경우에는 숫자형을 빌려온 문자형이라고 할 수 있다.

```r
kor <- 80
eng <- 81.67
mat <- 90.0
total <- kor + eng + mat # total = 251.67
avg <- total/3 #avg = 83.89
print(total)
print(avg)
```

```r
sgrade <- "A" #문자 "A"를 변수에 저장
mode(sgrade)
sname <- "홍길동" # 문자열 "홍길동"을 변수에 저장
mode(sname)
```

+ character는 문자형 데이터라는 의미

```r
cnum <- "1" # 문자 1 저장
mode(cnum)
num <- 1
mode(num)
```

+ ""를 쓰면 문자형으로 인식하고 데이터를 그냥 입력하면 숫자형으로 인식

### 논리형, 날짜형 및 시간형

```r
kor <- 95 # kor에 95 저장
above90 <- kor >=90 # kor 값에 90점 이상인가?
below90 <- kor < 90 # kor 값에 90점 미만인가?
print(above90)
print(below90)
```

```r
today1 <- Sys.Date() # 현재의 날짜 출력
print(today1)
today2 <- Sys.time() # 현재의 날짜와 시간 출력
print(today2)
```

### R언어 연산자

+ 보통 컴퓨터의 !는 NOT이라는 의미를 가지고 있다

```r
kor <- 80
above70 <- kor >=70 # kor가 70점 이상인가?
print(above70)
```

```r
kor <- 80
eng <- 70
above70 <- (kor >= 70) & (eng >= 70) # kor가 70점 이상이고
# eng가 70점 이상인가?
print(above70)
```

### 함수

+ 복잡한계산을 하나의 함수로 대체하여 사용한다.
+ 컴퓨터 프로그래밍에 필요한 다양한 기능을 함수 형태로 만들어 제공한다.
+ R언어는 다른 컴퓨터 언어보다 특정 문제를 해결할때 특정 함수로 짧게 해결할 수 있다.


### 패키지

+ 특정 문제 해결에 필요한 여러 함수들의 모음

### 수치 정보를 처리하는 기본 패키지 함수

```r
score <- c(80,90,70,65,55,30,60,90,88,100,76,30,55,61,89,68,78,70,88,82)
sum(score) # 반 합계 점수
mean(score) # 반 평균 점수
median(score) # 반 중앙값 점수
max(score) # 반 최고 점수
min(score) # 반 최저 점수
length(score) # 반 학생의 수
```

```r
score <- c(80,90,70,65,55,30,60,90,88,100,76,30,55,61,89,68,78,70,88,82)
sumScore <- sum(score)
avgScore <- mean(score)
cat("총점 = ", sumScore, "평균 점수 = ", avgScore)
```

```r
score <- c(80,90,70,65,55,30,60,90,88,100,76,30,55,61,89,68,78,70,88,82)
sumScore <- sum(score)
avgScore <- mean(score)
midScore <- median(score)
stdNum <- length(score)
maxScore <- max(score)
minScore <- min(score)

cat("총점 = ", sumScore, "평균 점수 = ", avgScore, "\n")
cat("중앙값 = ", midScore, "학생수 = ", stdNum, "\n")
cat("최대값 = ", maxScore, "최소값 = ", minScore, "\n")
```

+ "\n"은 줄을 바꾸라는 의미

### 강의 실습 과제

```r
sales <- c(80,90,70,65,55,30,60,90,88,100,76,30)
sumsales <- sum(sales)
avgsales <- mean(sales)
maxsales <- max(sales)
minsales <- min(sales)
smonth <- sort(sales)
Maxsales <- which.max(sales)
Minsales <- which.min(sales)

cat("판매총액 = ", sumsales, "월평균 판매금액 = ", avgsales, "\n")
cat("월 최대 판매금액 = ", maxsales, "월 최소 판매금액 = ", minsales, "\n")
cat("최대 판매한 달 = ", Maxsales, "최소 판매한 달 = ", Minsales, "\n")
print(smonth)
```
