### R언어의 조건 명령어

```r
inNUM <- readline("임의의 자연수를 입력하세요 : ")
number <- as.numeric(inNUM)
remainder <- number %% 2

if(remainder == 0){
    result <- "even number"
}else{
    result <- "odd number"
}
cat(number,"is an", result)
```

### R언어의 반복명령어

```r
for(n in 1:5){
    print(n)
}
```

```r
x <- c(1,3,5)
for (n in x){
    cat(n,"의 제곱",n*n,"\n")
}
```

```r
evenCount <- 0
oddCount <- 0
x <- c(3,4,9,1,6,2,6,7,3)
for (n in x)
    if( n %% 2 == 0){
        evenCount <- evenCount + 1
    }else{
        oddCount <- oddCount + 1
    }
cat("짝수의 개수 :",evenCount,"\n")
cat("홀수의 개수 :",oddCount,"\n")
```

```r
x <- 1; base <- 300
while (x <= base){
    x <- x*2
}
cat(base,"보다 크면서",base,"에 가장 가까운 2의 제곱승 값은",x,"이다","\n");
```

```r
x <- 0; base <- 100; multi <- 7;
while (x <= base){
    x <- x + multi;
}
cat(base,"을 초과하면서",base,"에 가장 가까운",multi,"의 배수는",x,"입니다")
```

```r
x <- 0; base <- 100; multi <- 7;
repeat{
    if(x > base) break
    x <- x + multi
}
cat(base,"을 초과하면서",base,"에 가장 가까운",multi,"의 배수는",x,"입니다","\n");
```

### 성적 데이터에서 등급과 순위 매기기

```r
id <- c("1","2","3","4")
name <- c("정우성","민철우","송중기","김영희")
kor <- c(80,60,90,70)
eng <- c(70,50,100,80)
mat <- c(95,70,95,80)
score1 <- data.frame(학번=id,이름=name,국어=kor,영어=eng,수학=mat)
print(score1)

avgValue <- apply(score1[,c(3:5)],1, mean)
print(avgValue)

finalGrade <- c()
for(n in avgValue){
    if(n>=90){
        Grade <- "A"
    } else if(n >= 80){
        Grade <- "B"
    } else if(n >= 70){
        Grade <- "C"
    } else if(n >= 60){
        Grade <- "D"
    } else{
        Grade <- "F"
    }
    finalGrade <- c(finalGrade,Grade)
}
print(finalGrade)
```

```r
score2 <- cbind(score1,평균 = avgValue, 등급 = finalGrade)
print(score2)
```

```r
score1 <- cbind(score1,평균 = avgValue, 등급 = finalGrade)
print(score1)
```

# 9장. 텍스트 데이터에서의 정보 찾기

### 텍스트 데이터의 특성

+ 정형화된 텍스트 데이터는 정해진 형식에 의해 정리되어 있으므로 통계 및 분석 함수 등을 이용해 정보를 쉽게 얻을 수 있다
+ 현재 정보화 시대에는 정리되어 있지 않은 텍스트들도 데이터로 간주하며 이들 데이터에서 정보를 추출하여 활용하고 있다 -> 비정형데이터
+ 원하는 정보를 얻기 위해서는 비정형데이터를 정형데이터로 변경하는 전처리 과정이 필요하다

### 텍스트 추출 함수

```r
str_including_pole <- grep("Pole",c("North Pole","South Pole","Equator"), value = TRUE)
print(str_including_pole)

index_of_pole <- grep("Pole",c("North Pole","South Pole","Equator"), value = FALSE)
print(index_of_pole)
```

+ value가 TRUE이면 Pole단어가 포함된 문자열만 추출하고 FALSE이면 Pole단어가 포함된 원소의 위치를 반환한다

```r
elec_code <- substr("LEDTV-48-B",1,5);
print(elec_code)
```

```r
list_word <- strsplit("7-24-2017",split="-");
print(list_word)
```

```r
vec_word <- unlist(list_word);
print(vec_word)
```

+ unlist는 리스트를 벡터로 변환시켜준다

```r
word <- strsplit("2017.7.24",split=".");
print(word)
word1 <- strsplit("2017.7.24",split=".",fixed=TRUE);
print(word1)
word2 <- strsplit("2017.7.24",split="\\.");
print(word2)
```

+ .은 일반적으로 정규표현식에서 임의의 한 문자를 가리킨다 그래서 저런 결과가 나타난다
+ 따라서 fixed나 \\.으로 표현해야 정상적으로 나타난다

```r
pword1 <- paste("2017","5","30");
print(pword1)
pword2 <- paste("2017","5","30",sep="-");
print(pword2)
```

```r
txt <- "Text Analytics is useful. Text Analytics is also interesting"
new_txt1 <- sub("Text","Data",txt);
print(new_txt1)
new_txt2 <- gsub("Text","Data",txt);
print(new_txt2)
```

+ 전체 데이터 중 하나만 바꾸고 싶을경우 sub 모두를 바꾸고 싶으면 gsub를 사용한다

```r
txt <- "Text-Analytics is useful.\n Text Analytics is also interesting.\n"
txt1 <- gsub("\n","",txt)
print(txt1)
txt2 <- gsub("-","",txt1)
print(txt2)
```

```r
txt <- "Text Analytics is useful. Text Analytics is also interesting"
small_txt <- tolower(txt);
print(small_txt)
capital_txt <- toupper(txt);
print(capital_txt)
```

```r
grade <- c("B","C","C","B","A","C","C")
freq_grade <- table(grade); print(freq_grade)
```
