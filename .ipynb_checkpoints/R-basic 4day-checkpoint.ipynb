### 편리한 상호작용과 파일 입출력

+ sprint() 함수는 cat()함수보다 유연하고 더 많은 기능을 가지고 있어 매우 자주 사용되는 출력 메세지를 생성하는 함수이다

```r
name <- readline("Input your name : ")
print(name)
```

```r
inwei <- readline("당신의 몸무게는?") #readline은 문자열로 들어간다
inhei <- readline("당신의 키는?")
wei <- as.numeric(inwei) # 입력한 문자열을 숫자값으로 변경
hei <- as.numeric(inhei)
vbmi <- (wei*10000)/(hei^2);
cat("당신의 BMI 지수는", vbmi,"입니다.\n")
```

```r
kor <- scan() # 숫자형 벡터를 생성할때 사용
print(kor)
```

```r
charvalue <- scan(what = character()) # 문자형 벡터를 생성할때 사용
print(charvalue)
```

```r
listvalue <- scan(what = list(name = character(),kor = numeric(),eng = numeric()))
print(listvalue) # 문자형과 숫자형으로 된 리스트를 생성할때 사용
```

```r
student1 <- data.frame() # 데이터 프레임 입력에 편리한 도구 edit 
student1 <- edit(student1) # 표를 생성할 때 사용
print(student1)
```

```r
name = "홍길동"
sum = 254
avg = sum / 3
outMsg <- sprintf("%s 학생의 총점은 %d점 이고, 평균은 %.1f입니다.\n",name, sum, avg)
cat(outMsg) # %s는 네임을 그 자리에 문자형으로 출력하라는 것 %d는 숫자형
```

```r
inwei <- readline("당신의 몸무게는?") #readline은 문자열로 들어간다
inhei <- readline("당신의 키는?")
wei <- as.numeric(inwei) # 입력한 문자열을 숫자값으로 변경
hei <- as.numeric(inhei)
vbmi <- (wei*10000)/(hei^2);
bmi <- sprintf("당신의 BMI 지수는 %.1f입니다.\n",vbmi)
cat(bmi)
```

```r
student1 <- data.frame() # 데이터 프레임 입력에 편리한 도구 edit 
student1 <- edit(student1) # 표를 생성할 때 사용
View(student1) # V는 대문자로 사용해야 에러가 안난다.
```

### 데이터 관리를 위한 파일 입출력

+ 상호작용을 위해 콘솔에서 입력받는 readline() 함수와는 L이 대문자고 끝에 s가 붙은 Lines이다

```r
setwd("D:/R with Big Data/강의 실습자료/ch_06")
txtLines <- readLines("ch6_movie_list.txt")
print(txtLines)
```

```r
setwd("D:/R with Big Data/강의 실습자료/ch_06")
txtLines <- readLines("ch6_movie_list.txt", n = 4) # 한 줄 단위로 4줄만 읽어오기
writeLines(txtLines, "ch6_movie_list_copy.txt") # 읽어온 4줄을 파일에 저장하기
print(txtLines)
```
