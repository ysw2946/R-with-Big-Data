### 데이터 묶음

+ 데이터 묶음으로써 많은 데이터를 편하게 정리하고 사용할 수 있다
+ 데이터를 묶음으로 관리하기 위해서는 하나의 이름을 사용해야 한다
+ 한가지의 데이터를 묶는방식은 속도측면에서 장점이 있다.
+ 여러 가지의 데이터를 묶어서 관리하는 방법은 편리성에 장점이 있다.
+ 하나의 열이나 행으로 묶는것을 벡터라고 하고 여러 행과 열을 묶으면 행렬이 된다.

```r
kor <- c(80, 60, 90, 70)
eng <- c(70, 50, 100)
korEng <- c(kor,eng)
math <- c(95, 70, "95", 80)
kor2 <- c(90,80, T, TRUE)
```

+ c는 combine으로 벡터를 만드는 함수이다
+ korEng는 두 벡터를 연결한 것으로 옆으로 연결되어진다
+ 데이터 묶음에 문자형이 하나라도 섞여있는 경우 모두 문자형으로 바뀐 벡터가 생성된다
+ T,TRUE는 숫자 1로 바뀌어서 숫자형 벡터가 생성된다

```r
kor <- 80
math <- 90
eng <- "B"
sco <- c(kor,math,eng)
tot <- sum(sco)
```

+ sco가 문자형 벡터이므로 문자형의 합은 구할 수 없기때문에 에러가 난다

### 특정 규칙이 있는 벡터 만들기

```r
x <- 1:5 # x는 1,2,3,4,5 의 5개의 원소를 갖는 벡터
cat("x : ",x, "\n")

y <- 10:5 # y는 10,9,8,7,6,5 의 6개의 원소를 갖는 벡터
cat("y : ",y,"\n")
```

```r
res1 <- seq(from = 1, to = 20, by = 2) # 1부터 20까지 2씩 증가
cat("res1 : ",res1, "\n")
res2 <- seq(1, 5, 0.5) #1부터 5까지 0.5씩 증가
cat("res2 : ",res2, "\n")
res3 <- seq(1, 10, length.out = 7) #1부터 10까지 7개 되는 자료 생성
cat("res3 : ",res3, "\n")
res4 <- seq(3, 1000, 3) #3부터 1000까지 3의 배수인 자료 생성
cat("res4 : ",res4, "\n")
```

### 행렬(Matrix)

+ 행렬은 2차원으로 데이터를 관리하며, 행이 가지는 의미와 열이 가지는 의미를 고려하여 데이터를 관리할 수 있다

```r
data = c(80, 60, 90, 70, 70, 50, 100, 80, 95, 70, 95, 80)
score <- matrix(data,nrow = 4)
print(score)
```

+ 행렬을 생성할때는 데이터는 벡터형으로 주어져야 한다
+ byrow 값을 안주면 열 우선으로 생성된다

```r
data = c(80, 60, 90, 70, 70, 50, 100, 80, 95, 70, 95, 80)
score <- matrix(data,ncol = 3, byrow = TRUE)
print(score)
```

```r
kor <- c(80, 60 , 90 ,70)
eng <- c(70, 50, 100, 80)
math <- c(95, 70, 95, 80)

scoreC <- cbind(kor, eng, math)
cat("cbind : \n "); print(scoreC)

scoreR <- rbind(kor, eng)
cat("rbind : \n"); print(scoreR)
```

+ 열방향으로 묶을건지 행방향으로 묶을 건지 정해야함 ex) rbind, cbind
+ 벡터를 가지고 행렬을 만들 수 있다

### 개별 값에 대한 접근과 연산(벡터)

```r
kor <- c(80, 60, 90, 70)
kor[3] <- 95
num <- kor[4]
evenStd <- kor[c(2,4)]
print(evenStd)
```

```r
kor <- c(80, 60, 90, 70)
eng <- c(70, 50, 100, 80)
math <- c(95, 70, 95, 80)
totStu <- kor + eng + math
cat("totStu : "); print(totStu)

res <- kor / math
cat("res : "); print(res)
```

###  개별 값에 대한 접근과 연산(행렬)

```r
score <- matrix(c(80, 60, 90, 70, 70, 50, 100, 80, 95, 70, 95, 80), nrow = 4)
print(score)
score[2,2] <- score[2,2] + 20
print(score)
```

```r
kor <- score[,1, drop = FALSE]
print(kor)
```

+ drop을 쓰게되면 벡터가 아니라 행렬로 만들어짐

```r
substu <- score[c(2,3),c(2,3)]
print(substu)
```

```r
midSco <- matrix(c(80, 60, 90, 70, 70, 50, 10, 80, 95, 70, 95, 80), nrow = 4)
finSco <- matrix(c(85, 75, 95, 80, 80, 70, 95, 90, 90, 85, 90, 85), nrow = 4)
midSco[3,2] <- 100
totalSco <- midSco + finSco
print(midSco)
print(finSco)
print(totalSco)
```

+ 행렬의 연산은 행과 열의 수가 같지 않으면 오류가 나고 연산이 안된다

```r
midSco <- matrix(c(80, 60, 90, 70, 70, 50, 10, 80, 95, 70, 95, 80), nrow = 4)
colnames(midSco) <- c("국어","영어","수학")
rownames(midSco) <- c("정우성","김철수","송중기","김영희")
print(midSco)
```
